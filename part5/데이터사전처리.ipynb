```python
df = pd.read_csv('./auto-mpg.csv', header=None)
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
print(df)
```

```python
mpg_to_kpl = 1.60934/3.78541 # == 0.425

df['kpl']=(df['mpg'] * mpg_to_kpl).round(2)
print(df)
```

```python
df.dtypes
```

```python
import numpy as np

print(df['horsepower'].unique()) # ? 는 NaN 이다 문자열을 numeric으로 바꾸기 전에 확인
df['horsepower'].replace('?', np.nan, inplace=True)
df.dropna(subset='horsepower', axis=0, inplace=True)
df['horsepower']=df['horsepower'].astype('float')
df.dtypes
```

```python
print(df['origin'].unique())
df['origin'].replace({1:'USA', 2:'EU', 3:'JPN'}, inplace=True)
print(df['origin'].unique())
print(df['origin'].dtypes)
```

```python
df['origin'] = df['origin'].astype('category')
df['model year'] = df['model year'].astype('category')
print(df['origin'].dtypes)
print(df.dtypes)
df['origin'] = df['origin'].astype('str')
print(df.dtypes)
```

```python
count, bin_dividers = np.histogram(df['horsepower'], bins=3)
print(count, bin_dividers)
bin_names = ['저출력', '보통출력', '고출력']
df['hp_bin'] = pd.cut(x=df['horsepower'],
                     bins=bin_dividers,
                     labels=bin_names,
                     include_lowest=True)
# print(df['hp_bin'])
# df['hp_bin'] = pd.cut(x=df['horsepower'],
#                      bins=bin_dividers,
#                      labels=bin_names)
print(df['hp_bin'])
```

```python
# 원핫 인코딩
horsepower_dummies = pd.get_dummies(df['hp_bin'])
horsepower_dummies
```

```python
from sklearn import preprocessing

label_encorder = preprocessing.LabelEncoder()
onehot_incoder = preprocessing.OneHotEncoder()

onehot_labeled = label_encorder.fit_transform(df['hp_bin'].head(15))
print(onehot_labeled)
print(type(onehot_labeled))
```

```python
onehot_reshaped = onehot_labeled.reshape(len(onehot_labeled), 1)
print(onehot_reshaped)
print(type(onehot_reshaped))
```

```python
onehot_fitted = onehot_incoder.fit_transform(onehot_reshaped)
print(onehot_fitted)
print(type(onehot_fitted))
print(onehot_fitted[0])
```

```python
# 데이터 정규화 max 값의 절대값으로 나눠 0 ~ 1 사이 범위로 값을 정규화 -범위가 더 클경우엔 못 씀

df['horsepower'].describe()
df['horsepower'] = df['horsepower']/abs(df['horsepower'].max())
print(df['horsepower'].describe())
```

```python
df = pd.read_csv('./auto-mpg.csv', header=None)

df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']

df['horsepower'].replace('?',np.nan,inplace=True)
df.dropna(subset=['horsepower'],axis=0,inplace=True)
df['horsepower']=df['horsepower'].astype('float')

min_x = df['horsepower'] - df['horsepower'].min()
print(min_x)
min_max = df['horsepower'].max() - df['horsepower'].min()
print(min_max)
df['horsepower']=min_x/min_max
print(df['horsepower'].describe())
```
