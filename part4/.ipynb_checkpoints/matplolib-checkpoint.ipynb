```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('./시도별 전출입 인구수.xlsx', engine='openpyxl', header=0)
df
```

```python
df = df.fillna(method='ffill') # NaN을 이전 정보로 채움
mask = (df["전출지별"] == '서울특별시') & (df["전입지별"] != '서울특별시')
print(mask)
df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul=df_seoul.rename({'전입지별':'전입지'}, axis=1)
```

```python
df_seoul.set_index('전입지', inplace=True)
```

```python
sr_one = df_seoul.loc['경기도']
plt.plot(sr_one.index, sr_one.values)
```

```python
plt.plot(sr_one.index, sr_one.values) # == plt.plot(sr_one)
plt.title('서울->경기 인구 이동')
plt.xlabel("기간")
plt.ylabel("이동인구수")
plt.show()
```

```python
plt.plot(sr_one.index, sr_one.values) # == plt.plot(sr_one)
plt.title('서울->경기 인구 이동')
plt.xlabel("기간")
plt.ylabel("이동인구수")
plt.show()
```

```python
plt.figure(figsize=(14, 5))
plt.xticks(rotation='vertical')
plt.plot(sr_one.index, sr_one.values) # == plt.plot(sr_one)
plt.title('서울->경기 인구 이동')
plt.xlabel("기간")
plt.ylabel("이동인구수")
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(14, 5))
plt.xticks(size=10, rotation='vertical') # == rotation=90 .... 45...0 참고
plt.plot(sr_one.index, sr_one.values, marker='o', markersize=10) # 마커 표시 추가
plt.title('서울->경기 인구 이동')
plt.xlabel("기간")
plt.ylabel("이동인구수")
plt.show()
print(plt.style.available) # 설정가능 스타일
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(14, 5))
plt.xticks(size=10, rotation='vertical') # == rotation=90 .... 45...0 참고
plt.plot(sr_one.index, sr_one.values, marker='o', markersize=10) # 마커 표시 추가
plt.title('서울->경기 인구 이동')
plt.xlabel("기간")
plt.ylabel("이동인구수")
plt.ylim(5000, 800000) # y축 범위 지정
plt.annotate('',
            xy=(20, 620000),    # 화살표 머리 부분(끝점)
            xytext=(2, 290000), # 화살표 꼬리 부분(시작점)
            xycoords='data',    # 좌표체계
            arrowprops=dict(arrowstyle='->', color='skyblue', lw=5) # 화살표 서식
            )

plt.annotate('',
            xy=(45, 430000),     # 화살표 머리 부분(끝점)
            xytext=(30, 580000), # 화살표 꼬리 부분(시작점)
            xycoords='data',     # 좌표체계
            arrowprops=dict(arrowstyle='->', color='olive', lw=5) # 화살표 서식
            )

plt.annotate('인구 이동 증가(1970-1995)', # 텍스트 입력
            xy=(10, 450000),              # 텍스트 위치 기준점
            rotation=25,                  # 텍스트 회전 각도
            va='baseline',                # 텍스트 상하 정렬
            ha='center',                  # 텍스트 좌우 정렬
            fontsize=15,                  # 화살표 서식
            )

plt.annotate('인구 이동 감소(1965-2017)', # 텍스트 입력
            xy=(40, 560000),              # 텍스트 위치 기준점
            rotation=11,                  # 텍스트 회전 각도
            va='baseline',                # 텍스트 상하 정렬
            ha='center',                  # 텍스트 좌우 정렬
            fontsize=15,                  # 화살표 서식
            )

plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(2, 1, 1)  # (행, 열, 부분)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(sr_one, 'D', markersize=10)
ax2.plot(sr_one, marker='o', markerfacecolor='green', markersize=10,
        color='olive', linewidth=2, label='서울->경기')
ax2.legend(loc='best')

ax1.set_ylim(5000, 800000)
ax2.set_ylim(5000, 800000)

ax1.set_title('서울->경기 인구 이동', size=20)
ax2.set_title('서울->경기 인구 이동', size=20)

ax1.set_xlabel("기간", size=12)
ax1.set_ylabel("이동인구수", size=12)

ax2.set_xlabel("기간", size=12)
ax2.set_ylabel("이동인구수", size=12)

ax1.set_xticklabels(sr_one.index, rotation=75)
ax2.set_xticklabels(sr_one.index, rotation=75)

ax1.tick_params(axis='x', labelsize=10)
ax1.tick_params(axis='y', labelsize=12)

ax2.tick_params(axis='x', labelsize=12)
ax2.tick_params(axis='y', labelsize=12)

plt.show()
```

```python
col_year = list(map(str, range(1970, 2017+1)))
df_3 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_year]
df_3
```

```python
plt.style.use('ggplot')

fig =plt.figure(figsize=(20, 5))
ax = fig.add_subplot(1, 1, 1)

ax.plot(col_year, df_3.loc['충청남도',:], marker='o', markerfacecolor='green', markersize=10,
        color='olive', linewidth=2, label='서울->충남')
ax.plot(col_year, df_3.loc['경상북도',:], marker='o', markerfacecolor='blue', markersize=10,
        color='skyblue', linewidth=2, label='서울->경북')
ax.plot(col_year, df_3.loc['강원도',:], marker='o', markerfacecolor='red', markersize=10,
        color='magenta', linewidth=2, label='서울->강원')

ax.legend(loc='best')

ax.set_title('서울->충남, 경북, 강원 인구 이동', size=20)

ax.set_xlabel("기간", size=12)
ax.set_ylabel("이동인구수", size=12)

ax.set_xticklabels(col_year, rotation=90)

ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)

plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(2, 2, 1)  # (행, 열, 부분)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.plot(col_year, df_3.loc['전라남도',:], marker='o', markerfacecolor='orange', markersize=10,
        color='yellow', linewidth=2, label='서울->전남') # 마커 표시 추가
ax2.plot(col_year, df_3.loc['충청남도',:], marker='d', markerfacecolor='green', markersize=10,
        color='olive', linewidth=2, label='서울->충남')
ax3.plot(col_year, df_3.loc['경상북도',:], marker='*', markerfacecolor='blue', markersize=10,
        color='skyblue', linewidth=2, label='서울->경북')
ax4.plot(col_year, df_3.loc['강원도',:], marker='+', markerfacecolor='red', markersize=10,
        color='magenta', linewidth=2, label='서울->강원')

ax1.set_ylim(5000, 80000)
ax2.set_ylim(5000, 80000)
ax3.set_ylim(5000, 80000)
ax4.set_ylim(5000, 80000)

ax1.set_title('서울->전남 인구 이동', size=15)
ax2.set_title('서울->충남 인구 이동', size=15)
ax3.set_title('서울->경북 인구 이동', size=15)
ax4.set_title('서울->강원 인구 이동', size=15)

# ax1.set_xlabel("기간", size=12)
# ax1.set_ylabel("이동인구수", size=12)
# ax2.set_xlabel("기간", size=12)
# ax2.set_ylabel("이동인구수", size=12)
# ax3.set_xlabel("기간", size=12)
# ax3.set_ylabel("이동인구수", size=12)
# ax4.set_xlabel("기간", size=12)
# ax4.set_ylabel("이동인구수", size=12)

ax1.set_xticklabels(col_year, rotation=75)
ax2.set_xticklabels(col_year, rotation=75)
ax3.set_xticklabels(col_year, rotation=75)
ax4.set_xticklabels(col_year, rotation=75)

ax1.tick_params(axis='x', labelsize=12)
ax1.tick_params(axis='y', labelsize=12)
ax2.tick_params(axis='x', labelsize=12)
ax2.tick_params(axis='y', labelsize=12)
ax3.tick_params(axis='x', labelsize=12)
ax3.tick_params(axis='y', labelsize=12)
ax4.tick_params(axis='x', labelsize=12)
ax4.tick_params(axis='y', labelsize=12)

ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best')

plt.show()
```

```python
cols = df.columns[2:].astype(int)
df_4 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], :]

plt.style.use('ggplot')
df_4.columns = cols
tdf_4 = df_4.T

tdf_4.plot(kind='area', stacked=False, alpha=0.2, figsize=(20, 10))

plt.title('서울->타시도 인구 이동')
plt.xlabel("기간")
plt.ylabel("이동인구수")
plt.legend(loc='best', fontsize=15)

plt.show()
```

```python
plt.style.use('ggplot')
df_4.columns = cols
tdf_4 = df_4.T

tdf_4.plot(kind='area', stacked=True, alpha=0.2, figsize=(20, 10))

plt.title('서울->타시도 인구 이동')
plt.xlabel("기간")
plt.ylabel("이동인구수")
plt.legend(loc='best', fontsize=15)

plt.show()
```

```python
fig = plt.figure(figsize=(20, 5))

ax = tdf_4.plot(kind='area', stacked=True, alpha=0.2, figsize=(20, 10))

ax.set_title('서울->타시도 인구 이동', size=30, color='brown', weight='bold')
ax.set_xlabel("기간", color='blue')
ax.set_ylabel("이동인구수", color='blue')
ax.legend(loc='best', fontsize=15)

plt.show()
```

```python
df_5 = tdf_4.loc[2010:2017+1, :]
df_5.plot(kind='bar', stacked=False, alpha=0.2, figsize=(20, 10))

plt.title('서울->타시도 인구 이동')
plt.xlabel("기간")
plt.ylabel("이동인구수")
plt.legend(loc='best', fontsize=15)

plt.show()
```

```python
df_6 = df_seoul.loc[['충청남도', '강원도', '경상북도', '전라남도'], '2010':'2017']
cols = df_6.columns[:].astype(int)
df_6.columns = cols

df_6['합계'] = df_6.sum(axis=1)
df_total = df_6[['합계']].sort_values(by='합계', ascending=True)
df_total.plot(kind='barh', color='cornflowerblue', width=0.5, figsize=(10, 5))

plt.style.use('ggplot')
plt.title('서울->타시도 인구 이동')
plt.xlabel("이동인구수")
plt.ylabel("전입지")
plt.legend(loc='best', fontsize=15)

plt.show()
```
