```python
import folium

seoul_map = folium.Map(location=[37.55, 126.98], zoom_start=12)
seoul_map.save('./seoul_map.html')
```

```python
import pandas as pd

df = pd.read_excel('./서울지역 대학교 위치.xlsx', engine='openpyxl')
df = df.rename(columns={'Unnamed: 0':'대학명'})
df.set_index('대학명', inplace=True)
seoul_map = folium.Map(location=[37.55, 126.98], zoom_start=12)

for name, lat, alt in zip(df.index, df.위도, df.경도):
    folium.Marker([lat, alt], popup=folium.Popup(name, max_width=500)).add_to(seoul_map)
    
seoul_map
```

```python
seoul_map1 = folium.Map(location=[37.55, 126.98], zoom_start=12)

for name, lat, alt in zip(df.index, df.위도, df.경도):
    folium.CircleMarker([lat, alt], 
                        radius=10, 
                        color='brown', 
                        fill=True,
                        fill_color='coral',
                        fill_opacity=0.7,
                        popup=folium.Popup(name, max_width=500)).add_to(seoul_map1)
    
seoul_map1
```

```python
import json

# 년도별 경기도 인구 정보를 읽는다
df_pop = pd.read_excel('경기도인구데이터.xlsx', index_col='구분', engine='openpyxl') # index를 '구분' column의 데이터로 읽음
df_pop.columns=df_pop.columns.map(str)  # 컬럼 명이 년도(정수)로 되어 있어 문자열로 변환

# 경기도 행정구역 경계 json을 읽어 dictionary로 반환
try:
    geo_data = json.load(open('경기도행정구역경계.json', encoding='utf-8'))
except:
    geo_data = json.load(open('경기도행정구역경계.json', encoding='utf-8-sig'))
    
# print(geo_data)

g_map = folium.Map(location=[37.5502, 126.982], zoom_start=9)

# year = '2007'
year = '2017'

folium.Choropleth(geo_data=geo_data,         # 지도경계
                  data=df_pop[year],         # 표시데이터
                  columns=[df_pop.index, df_pop[year]], # 열 지정
                  fill_color='YlOrRd',
                  fill_opacity=0.7,
                  line_opacity=0.3,
                  threshold_scale=[1000, 100000, 300000, 500000, 700000],
                  key_on='feature.properties.name'
                 ).add_to(g_map)

g_map
```
